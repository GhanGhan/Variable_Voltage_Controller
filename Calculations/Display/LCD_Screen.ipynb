# NHD-C12832A1Z-FSW-FBW-3V3
- COG (Chip-On-Glass) Liquid Crystal Display Module
- 128 x 32 Pixels
- 3.0V LCD, 3.0V Backlight
    - Supply Voltage $2.7V <=V_{DD} <= 3.3V$
    - Backlight supply voltage $2.9V <=V_{LED} <= 3.1V$
    - Backlight supply Current (For $V_{LED}= 3.0V$) $10mA <=I_{LED} <= 36mA$
- 4-wire SPI Interface
- 
- **$V_{LCD}$ which is nominally 6.0V, is internally generated by the LCD module's driver IC, the ST7565R Controller, which has a built-in charge pump or voltage booster to genertate the necessary voltage, which is why it doesn't appear in the pinout or wiring diagram
