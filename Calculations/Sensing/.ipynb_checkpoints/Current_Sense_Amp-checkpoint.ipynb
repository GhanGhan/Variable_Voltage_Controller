# Purpose
- Convert sensed current to a voltage value between 0V and 3.3V

#  INA215
- low or high side measurment, bidirectional zero-drift, current shunt motors
- wide common-mode range: -0.3V to 26V
- Gain: 75 V/V
- Gain Error (max over temperature): 0.5% for veresion C, 1% for versions A and B
- Offset Voltage 65uV max
    - typically 1uV
- max quiescent current: 100uA
- 
- typical CMMR (Common-mode rejection ratio): 140 dB
- Swing to V+ supply rail:  (V+) – 0.05
- Swing to GND: (VGND) + 0.005
- **Full-scale range**: maximum voltage drop accross the shunt resistor that the current-sense amplifier can measure accurately within its specified range
    - if the offset voltage is low, then the design can still maintain high accuracy with a low full-scale shunt voltage
    - shunt resistor can have a lower value, less power will be dissipated
