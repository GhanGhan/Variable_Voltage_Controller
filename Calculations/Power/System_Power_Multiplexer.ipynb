# Purpose
- Choose between powering the circuit from a DC power Jack or a (DC) screw terminal
- Will be used to power the 3.3V Buck Converter and the 1-20V Buck-Boost Converter
- There is no priority, if both power sources are connected, which ever one possesses the higher voltage will be chosen
- Must protect the system from reverse polarity and reverse current conditions

## Implementation - ORing Power Supplies with Ideal Diodes

Figure 8 of Ideal Diodes Application Note
- LM74700-Q1 combined with external NMOS transistors used in ORing solution
- forward diode voltage drop is reduced as the external NMOS is turned on during normal operation
- LM74700-Q1 quickly detects the reverse current and quickly pulls down the NMOS gate, leaving the body iode of the MOSFET to block reverse current flow
  - it constantly senses the voltage difference between the annode and cathode pins
  - a fast comparator shuts down the gate drive within $0.75 \micro s$ as soon as $V_{anode}-V_{cathode} < 11mV$
  - it turns on the gate once the $V_{anode}-V_{cathode} > 50mV$
