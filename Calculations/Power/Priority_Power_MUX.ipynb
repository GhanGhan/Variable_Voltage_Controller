# Discrete Automatic Power MUX with Priority

- Implemented using 4 discrete PMOS transistors, 2x paths of back to back FETS 
- Uses a comparator followed inverter logic to ensure only onse set of switches is turned on at  a time

# Purpose 
- Choose between powering all 3V3 circuitry from either the Buck converter or 3V3V STLink pin
    - The buck converter has priority, meaning if both power sources are available, the Buck converter will be chosen

# MCP6541T-I/OT 
- **Comparator**
- Supply voltage: 1.6V to 5.5V
- Rail-to Rail Input: Vss - 0.3V to VDD + 0.3V
- Internal hysteries of 3.3mV (typ) eliminates output switching due to internal input noise voltage
- Max output current: 30mA
- Bypass Capacitors: 0.01uF to 0.1uF
- Quiescent Current: 0.6uA (typ), 1.0 uA (max)

# BAT54WS-7-F
- **SCHOTTKY BARRIER DIODE**
- Reverse Breakdown Voltage: 30V
- Forward Voltage: Max of 240mV for a Forward current of 0.1mA
- Reverse Leakage Current: 2uA for a Reverse voltage of 25V
- *This is used to choose the power supply for the comparator, as it is expected to draw less than 1uA, not worried about conduction losses*

# AO3416
- **NMOS Transistor**
- Max Drain-Source Votlage: 20V
- Max Gate Source Voltage: 8V
- Max Continous Drain Voltage: 6.5A
- On-State Resistance (VGS = 4.5V) < 22m ohms

# AO3416
