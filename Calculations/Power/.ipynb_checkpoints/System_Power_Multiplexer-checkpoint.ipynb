# Purpose
- Choose between powering the circuit from a DC power Jack or a (DC) screw terminal
- Will be used to power the 3.3V Buck Converter and the 1-20V Buck-Boost Converter
- There is no priority, if both power sources are connected, which ever one possesses the higher voltage will be chosen
- Must protect the system from reverse polarity and reverse current conditions

## Implementation - ORing Power Supplies with Ideal Diodes

Figure 8 of Ideal Diodes Application Note
- LM74700-Q1 combined with external NMOS transistors used in ORing solution
- forward diode voltage drop is reduced as the external NMOS is turned on during normal operation
- LM74700-Q1 quickly detects the reverse current and quickly pulls down the NMOS gate, leaving the body iode of the MOSFET to block reverse current flow
  - it constantly senses the voltage difference between the annode and cathode pins
  - a fast comparator shuts down the gate drive within $0.75 \micro s$ as soon as $V_{anode}-V_{cathode} < 11mV$
  - it turns on the gate once the $V_{anode}-V_{cathode} > 50mV$

#  LM74700D-Q1
## Overview
- automotive AEC Q100 qualified ideal diode controller
- used for reverese polarity protection with external NMOS and for OR-ing configuration
- 3.2V to 65V input range (3.9V start up)
- 20mV **anode** to **cathode** forward voltage drop regulation
- enables turnoff off of the MOSFET during reverse current events, fast $< 0.75 \micro s$ response to reverse current dropping
    - ensures ZERO DC reverse current flow
- Voltage drop across FET continuously monitored between ANODE and CATHODE pins, the source and drain of the NMOS transistor,
    - GATE to ANODE voltage is adjusted as needed to regulate the forward voltage drop at 20mV
- closed loop regulation scheme enabled turn off of the FET during a reverse current event and make sure zero DC reverse current flows
    - reverse current condition detected voltage across ANODE and CATHODE reduces to below 11mV
    - Gate pin will then be internally connected to the ANODE pin, turninng off the MOSFET
    - body diode used to block any reverse current
- internal charge pump is used to drive external NMOS to a maximimujm gate drive voltage of 15V
- VCAP is the external charge pump capacitor
- Gate driver used to control the NMOS by setting the **Gate-Anode** voltage to the corresponding mode of operation
- **Gate driver is enabled when:** $V_{EN} > V_{EN_{IL}}$, VCAP-ANODE  > Undervoltage lockout voltage = 6.6V, ANODE voltage >  ANODE POR threshold = 2.8V


### Shutdown Mode
- occurs when the EN pin voltage is below the specified input low threshold $V_{EN_{IL}}$
- both the gate driver and charge pump are disabled
- **Forward curent flow through the NMOS is not interrupted, but conducted via the body iode**
### Conduction Mode
- Occurs when the gate driver is enabled
- 3 Regions: *Regulated Conduction Mode, Full Conduciton Mode, and Reverse Current Protection Mode*
#### Regulated Conduction Mode
- gate driver must be enabled, and current from the source to drain drain of the NMOS transistor must be within the range to result in the ANODE to CATHODE voltage drop of -11mV to 50mV
- i.e. $I_{DS}R_{DSON} <= 50mV$
#### Full Conduction Mode
- gate driver must be enabled, and the current from the source to drain of the external NMOS transistor must be large enough to result in an ANODE to CATHODE voltage drop that is greater than 50mV
- the GATE to ANDOE voltage will be approximately the same as the VCAP to ANODE voltage as the GATE and VCAP will be connected
- the external transistors on resistance in minimized, resulting in reduced power loss when forward currents are large
### Reverse Current Protection Mode
- gate driver must be enabled
- CATHODE voltage is at least 11mV higher than ANODE voltage
- GATE becomes internally connected to ANODE pin, disabling the FET
## Design Procedure
### Mosfet selection
- important parameters are the:
    - *max continous drain current*, must be higher than the max continuous load current
    - *maximum drain to source voltage*, must be able to handle voltage dif between multiplexer input and output
      - TI recommends MOSFETS with voltage rating up to 60V
    - *maximum source current through the body diode*, during startup, inrush current flows through the body diode to charge the bulk hold-up capacitors, the maximum source current through the body diode must be higher than the inrush current that can be seen in the application
    - *drain-to-source on resistance*:
        -  Lowest possible value good for reducing MOSFET conduction loss
      - Higher value good for reverse current detection
      - TI suggest to choose $R_{DS(ON)}$ such that at nominal operating current, the forward voltage drop is close to 20mV and no more than 50mV
      - Choose a gate threshold voltage of 2V to 2.5V max, this FET will have a low turn-ON time
    - max $V_{GS}$ this IC can drive is 15V, so chose a mosfet with a $V_{GS}$ max that is at least 15V
- **Chose the IRFML8244TRPBF NMOS transistor**
### Capacitors:
- VCAP: min of 0.1uF, or greater than $10C_{ISS}$ of the MOSFET
- $C_{IN} >= 22nF$, $C_{IN} >= 100nF$
