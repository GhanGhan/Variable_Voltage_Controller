# Overview
- ARM Cortex M0+ 32-bit MCU
- 17 I/Os
- Voltage supply: 2V to 3.6V
- Speed: 4 to 48 MHz crystal oscillator
- RAM: 8Kb
- Flash Memory: 64Kb
- 12bit ADC
- Communication interfaces: I2C supporting Fast-mode plus (1 Mbit/s), USARTs, SPIs (32Mbit/s)
- Development support: SWD
