# Microcontroller ADC Overview
- native 12 bit analog-to-digital converter
- can perform conversions in a **single-shot** or **scan** mode
    - in scan mode, automatic conversion performed on a selected group of analog inputs
- ADC frequency independent from CPU frequency, allowing for max sampling rate at low CPU speeds
- auto-shutdown guarantees that ADC is powered off except during auto-conversion phase
- can be served by the DMA controller
- features a hardware oversampler up to 256 samples, improving resolution to 16 bits (refer to AN2668)
- an analog watchdog feature allows for monitoring of the converted voltage of 1, some or all scanned channels
    - an interrupt is generated when the converted volage is outside the programmed thresholds
- events generated by general-purpose timers can be internally connected to the ADC start triggers

# Analog-to-Digital Characteristics
- minimum clock frequency of 140kHz, maximum clock frequency of eitehr 16MHZ or 35MHz
- max sampling rate at 12 bits: 2.50MSps
- max external trigger frequency = clock frequency/15 (assuming 12-bit resolution)
- max external input impedance: 50kohm (sampling cycle of 160.5 at 35MHz)
- internal sample and hold capacitor: 5 pF
- ADC power-up time: 2 conversion cycles
- Calibration time 2.35us (assuming clock frequency of 35MHz)

- Max ET: 6.5 LSB
- Max EO: 4.5 LSB
- Max EG: 5 LSB
- Max EL: 3 LSB
